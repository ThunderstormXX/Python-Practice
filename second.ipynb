Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
import math

class Vector:
    def __init__(self, vector_values_list):
        if not isinstance(vector_values_list, list):
            raise ValueError("Ожидаемый формат : list")
        if len(vector_values_list) == 0:
            raise ValueError("Ожидалось, что list не пустой")
        self.values = vector_values_list
    def __repr__(self):
        return f"Vector({self.values})"
    def __len__(self):
        return len(self.values)
    def __getitem__(self, i):
        return self.values[i]
    def __setitem__(self, i, value):
        self.values[i] = value
    def __add__(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Складывать можно только Vector с Vector")
        if len(self) != len(other):
            raise ValueError("Не соответствуют длины двух Vector")
        return Vector([self[i] + other[i] for i in range(len(self))])

    def __sub__(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Вычитать можно только Vector из Vector")
        if len(self) != len(other):
            raise ValueError("Не соответствуют длины двух Vector")
        return Vector([self[i] - other[i] for i in range(len(self))])

    def __mul__(self, other):
        if isinstance(other, Vector):
            if len(self) != len(other):
                raise ValueError("Не соответствуют длины двух Vector")
            return sum([self[i] * other[i] for i in range(len(self))])
        elif isinstance(other, (int, float)):
            return Vector([other * x for x in self.values])
        else:
            raise ValueError("Вектор можно умножить только на число или вектор")
    
    def __rmul__(self, other):
        return self.__mul__(other)

    def angle_cos(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Ожидается Vector")
        if len(self) != len(other):
            raise ValueError("Ожидается ,что Vector имеет такой же размер")
        dot_product = self * other
        norm1 = math.sqrt(sum([x ** 2 for x in self.values]))
        norm2 = math.sqrt(sum([x ** 2 for x in other.values]))
        if norm1 == 0 or norm2 == 0:
            raise ValueError("Ожидается , что один из Vector не нулевой")
        return dot_product / (norm1 * norm2)

    def norm(self):
        return math.sqrt(sum([x ** 2 for x in self.values]))

a = Vector([1,2,0])
b = Vector([1,2,1])
print( a + b ," " , a - b , " " , a*b , " ", a*b*a ," ",a*10, " " , a.angle_cos(b) , " " , a.norm() )
print(len(a) , " " , a[2] )
a[2] = 5
print(a[2])
````

````python
def copy_class_attrs(cls):
    class new(cls):
        def __init__(self, *args, **kwargs):
            super().__init__()
            for name in dir(*args):
                if not name.startswith('_'):
                    if hasattr(self, name):
                        setattr(self, name + "_new", getattr(*args, name))
                    else:
                        setattr(self, name, getattr(*args, name))
    return new


@copy_class_attrs
class Class:
    attr1 = 1
    attr2 = 2 
    attr3 = 3

# От которого будем брать атрибуты
class Class2:
    attr1 = 11
    attr2 = 22
    _attr3 = 33

obj = Class(Class2)

print(obj.attr1)
print(obj.attr2)
print(obj.attr3)
print(obj.attr1_new)
print(obj.attr2_new)
# print(obj._attr3) # 'NewCls' object has no attribute '_attr3'
````

````python
class SaveMixin:
    def save(self, path):
        with open(path, 'w') as f:
            data = str(self.data)
            f.write(data)      
class LoadMixin:
    def load(path, cls):
        with open(path, 'r') as f:
            data = f.read()
            pass
        #Не успел
class CopyMixin:
    def copy(self):
        data = self.data.copy()
        return Matrix(data)

class Matrix(SaveMixin, LoadMixin , CopyMixin):
    def __init__(self, *args, default=0):
        if len(args) == 1 and isinstance(args[0], list):
            self.data = args[0]
            self.rows = len(args[0])
            self.cols = len(args[0][0])
        elif len(args) == 1 and isinstance(args[0], int):
            self.data = [[default] * args[0]]
            self.rows = 1
            self.cols = args[0]
        elif len(args) == 2 and all(isinstance(arg, int) for arg in args):
            self.data = [[default] * args[1] for _ in range(args[0])]
            self.rows = args[0]
            self.cols = args[1]
        elif len(args) == 3 and all(isinstance(arg, int) for arg in args):
            self.data = [[default] * args[1] for _ in range(args[0])]
            self.rows = args[0]
            self.cols = args[1]
        else:
            raise TypeError("Invalid arguments")
    
    def __getitem__(self, index):
        if isinstance(index, int):
            return self.data[index]
        elif isinstance(index, tuple):
            row, col = index
            return self.data[row][col]
        else:
            raise TypeError('Invalid argument type')
    
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            self.data[index[0]][index[1]] = value
        else:
            self.data[index] = value
    
    def __iter__(self):
        for i in range(self.rows):
            yield self.data[i]
    
    def __add__(self, other):
        if not isinstance(other, Matrix) or self.rows != other.rows or self.cols != other.cols:
            raise ValueError("Invalid operand")
        result = Matrix(self.rows, self.cols)
        for i in range(self.rows):
            for j in range(self.cols):
                result[i, j] = self[i, j] + other[i, j]
        return result
    
    def __sub__(self, other):
        if not isinstance(other, Matrix) or self.rows != other.rows or self.cols != other.cols:
            raise ValueError("Invalid operand")
        result = Matrix(self.rows, self.cols)
        for i in range(self.rows):
            for j in range(self.cols):
                result[i, j] = self[i, j] - other[i, j]
        return result
    
    def __mul__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            result = Matrix(self.rows, self.cols)
            for i in range(self.rows):
                for j in range(self.cols):
                    result[i, j] = self[i, j] * other
            return result
        elif isinstance(other, Matrix) and self.cols == other.rows:
            result = Matrix(self.rows, other.cols)
            for i in range(self.rows):
                for j in range(other.cols):
                    for k in range(self.cols):
                        result[i, j] += self[i, k] * other[k, j]
            return result
        else:
            raise ValueError("Invalid operand")

    def __rmul__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            return self.__mul__(other)
        else:
            raise ValueError("Invalid operand")
    def __str__(self):
        return "Matrix([" + '],['.join([' ,'.join([str(cell) for cell in row]) for row in self.data]) + "])"

    def dot(self, other):
        return self.__mul__(other)
    
    def transpose(self):
        result = Matrix(self.cols, self.rows)
        for i in range(self.rows):
            for j in range(self.cols):
                result[j, i] = self[i, j]
        self = result
        return self

    def trace(self):
        if self.rows != self.cols:
            raise ValueError('Matrix is not square')

        result = 0
        for i in range(self.rows):
            result += self[i, i]
        return result
    
    def determinant(self):
        if self.rows != self.cols:
            raise ValueError("Matrix is not square")
        if self.rows == 1:
            return self[0, 0]
        elif self.rows == 2:
            return self[0, 0] * self[1, 1] - self[0, 1] * self[1, 0]
        else:
            result = 0
            for i in range(self.rows):
                minor = Matrix([row[:i] + row[i+1:] for row in self.data[1:]])
                result += (-1)**i * self[0, i] * minor.determinant()
            return result
````
